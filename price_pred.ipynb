Data Cleaning

```python
df2 = df1.drop(['area_type','society', 'balcony', 'availability'], axis=1)
df2.head()
```

```python
# df2.bath = df2.bath.fillna(df2.bath.median())
df3 = df2.dropna()
df3.isnull().sum()
```

```python
df3['bhk'] = df3['size'].apply(lambda x: int(x.split(' ')[0]))
df3.head()
```

```python
df3[df3.bhk>20]
```

```python
df3.total_sqft.unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df3[~df3['total_sqft'].apply(is_float)]
```

```python
def con_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return float(tokens[0])+float(tokens[1])/2
    try:
        return float(x)
    except:
        return None
```

```python
con_sqft_to_num('2166')
```

```python
df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(con_sqft_to_num)
df4.head()
```

```python
df4.loc[30]
```

Feature Engineering

```python
df5 = df4.copy()
```

```python
df5['price_per_sqft'] = df5['price']/df5['total_sqft']
df5.head()
```

```python
df5.location.unique() #dimensionality curse
```

```python
df5.location = df5.location.apply(lambda x:x.strip())
```

```python
location_stat = df5.groupby('location')['location'].count().sort_values(ascending=False)
location_stat
```

```python
loc_less_than_10=location_stat[location_stat<=10]
len(df5.location.unique())
```

```python
df5.location = df5.location.apply(lambda x:'other' if x in loc_less_than_10 else x)
```
